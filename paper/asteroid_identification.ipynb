```python
import warnings

warnings.filterwarnings("ignore", module="erfa")

import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.time import Time

from jorbit.mpchecker import mpchecker

mpchecker(
    coordinate=SkyCoord(97.36206 * u.deg, 15.65564 * u.deg),
    time=Time("2023-11-23"),
    radius=5 * u.arcmin,
    extra_precision=True,
    extra_precision_gravity="default solar system",
    observer="Kitt Peak",
)
```

```python
m = mpchecker(
    coordinate=SkyCoord(97.36206 * u.deg, 15.65564 * u.deg),
    time=Time("2023-11-23"),
    radius=5 * u.arcmin,
    extra_precision=True,
    extra_precision_gravity="default solar system",
    observer="Kitt Peak",
)
print(SkyCoord(m["ra"], m["dec"], unit="deg").to_string("hmsdms", sep=":"))
print(SkyCoord(97.36206 * u.deg, 15.65564 * u.deg).to_string("hmsdms", sep=" "))
```
